```python
url1="https://data.covid19india.org/v4/min/data.min.json"
url2="https://data.covid19india.org/v4/min/timeseries.min.json"
```

```python
x=requests.get(url1)
```

```python
xn=json.loads(x.content)
xn['AN']
```

```python
d1= {'state':[],
     'delta_tested':[],
     'delta_vaccinated1':[],
     'delta_vaccinated2':[],
     'delta7_confirmed':[],
     'delta7_recovered':[],
     'delta7_tested':[],
     'delta7_vaccinated1':[],
     'delta7_vaccinated2':[],
     'population':[],
     'total_confirmed': [],
  'total_deceased': [],
  'total_recovered': [],
  'total_tested': [],
  'total_vaccinated1': [],
  'total_vaccinated2': []}
```

```python
for state in xn.keys():
  xn1=xn[state]
  d1['state'].append(state)
  for varient in ['delta']:
    xn2=xn1.get(varient)
    for status in ['tested','vaccinated1', 'vaccinated2']:
      if xn2 is None:
        value=None
      else:
        value=xn2.get(status)
      key= varient+"_"+status
      d1[key].append(value)
  for varient in ['delta7']:
    xn2=xn1.get(varient)
    for status in ['confirmed','recovered','tested','vaccinated1', 'vaccinated2']:
      if xn2 is None:
        value=None
      else:
        value=xn2.get(status)
      key= varient+"_"+status
      d1[key].append(value)
  for varient in ['meta']:
    xn2=xn1.get(varient)
    for status in ['population']:
      if xn2 is None:
        value=None
      else:
        value=xn2.get(status)
      key= "population"
      d1[key].append(value)
  for varient in ['total']:
    xn2=xn1.get(varient)
    for status in ['confirmed','deceased','recovered','tested','vaccinated1','vaccinated2']:
      if xn2 is None:
        value=None
      else:
        value=xn2.get(status)
      key= varient+"_"+status
      d1[key].append(value)
```

# 1. Weekly evolution of number of confirmed cases, recovered cases, deaths, tests. For instance, your dashboard should be able to compare Week 3 of May with Week 2 of August 

```python
ydf['dates']
```

```python
ydf['dates'] = pd.to_datetime(ydf['dates'])
```

```python
ydf['dates']
```

```python
gdf=ydf.groupby(ydf['dates'].dt.week).agg({'delta_confirmed':'sum', 'delta_recovered':'sum', 'delta_tested':'sum','delta_deceased':'sum'})
gdf
```

```python
plt.figure()
gdf.plot(figsize=(10,8))
plt.xlabel("week", size=16)
plt.ylabel("Number of cases", size=16)
plt.title("Number of cases over the weeks", size=20)
plt.show()
```

# 2. Let’s call testing ratio(tr) = (number of tests done) / (population), now categorise every district in one of the following categories:-

- Category A: 0.05 ≤ tr ≤ 0.1
- Category B: 0.1 < tr ≤ 0.3
- Category C: 0.3 < tr ≤ 0.5
- Category D: 0.5 < tr ≤ 0.75
- Category E: 0.75 < tr ≤ 1.0
Now perform an analysis of number of deaths across all category.


```python
xdf.columns
```

```python
xdf['testing ratio']=xdf['total_tested']/xdf['population']
```

```python
xdf
```

```python
category = []
for cat in xdf['testing ratio']:
    if 0.05<=cat<=0.1:    
      category.append('A')
    elif 0.1<cat<=0.3:  
      category.append('B')
    elif 0.3<cat<=0.5:  
      category.append('C')
    elif 0.5<cat<=0.75: 
      category.append('D')
    else:  
      category.append('E')
```

```python
xdf['category']=category
```

```python
xdf
```

```python
xdf.columns
```

```python
xdf.to_excel("covid_19.xlsx")
xdf.to_csv("Covid_19.csv")
```

```python
deceased_by_category=xdf.groupby('category').agg({'total_deceased':'sum'})
```

```python
deceased_by_category
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(data=deceased_by_category, y='total_deceased', x='category')
plt.xlabel("category", size=15)
plt.ylabel("Number of deceased", size=15)
plt.title("category Vs deceased", size=18)

```
